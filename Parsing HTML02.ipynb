### Scrappy

```python
from scrapy.crawler import CrawlerProcess  #scrappy in jupyter notebook
import scrapy


class SeekSpider(scrapy.Spider):
    name = "job_search"                 #spider name
    allowed_domains = ["seek.com.au"]   #web page
    
    #seed with the seek landing site for all data jobs in AUS
#     start_urls = ['https://www.seek.com.au/data-scientist-jobs/in-All-Sydney-NSW']
    start_urls = ['https://www.seek.com.au/data-scientist-jobs']
    
    custom_settings = {
        'FEED_FORMAT' : 'csv',
        'FEED_URI' : 'seek_test15.csv'
        }
    
    def parse(self, response):
        
        print("-----------------")
        print("I just visited :" + response.url)
        print("-----------------")
        
        #grab list of all job
        urls = response.css('article') # Website has changed elements, Anita changed this to 'article'
        
        
        for job in urls: # 
            # fetch duties and tasks in the li tags
            list_of_dutiestasks = []
            
            duties_list = job.css('ul li')
            for each_duty in duties_list:
                list_of_dutiestasks.append(each_duty.css('span::text').extract_first())
           
            # initialize joblist
            joblist = {}
#             jobCompany = {}
            
            # fetch job elements 
            joblist['job_title'] = job.css('article::attr(aria-label)').extract_first()
            joblist['jobCompany'] = job.css('a[data-automation="jobCompany"]::text').extract_first()
            joblist['jobLocation'] = job.css('a[data-automation="jobLocation"]::text').extract_first()
            joblist['jobSalary'] = job.css('span[data-automation="jobSalary"]::text').extract_first()
            joblist['jobShortDescription'] = job.css('span[data-automation="jobShortDescription"]::text').extract_first()
            
            
#             joblist['dotPoints'] = job.css('ul[class="yvsb870 yvsb873 _1sx92fk0 _1sx92fk4"]::text').extract_first()
#             data-automation="jobCompany"
# data-automation="jobLocation"
#  ul class yvsb870 yvsb873 _1sx92fk0 _1sx92fk4
# span data-automation="jobShortDescription"
# data-automation="jobSalary"
            
#             yield joblist # shouldn't be there, already yeilded leter 
            
                    
                
        # grab the link for the next page
        next_page = response.css('a[data-automation="page-next"]::attr(href)').extract_first()
        
        # if 'next page' link present
        if next_page is not None:
            # join next page link with Domain to ensure correct operation
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback = self.parse)
    

        
process = CrawlerProcess()
process.crawl(SeekSpider)
process.start()
```
