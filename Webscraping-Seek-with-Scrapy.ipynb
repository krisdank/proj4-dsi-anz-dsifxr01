```python
from scrapy.crawler import CrawlerProcess  #scrappy in jupyter notebook
import scrapy


class JobSpider(scrapy.Spider):

    custom_settings = {
    'FEED_FORMAT' : 'json',
    'FEED_URI' : 'job.json'
    }
        
    name = "jobseek-new"
    allowed_domains = ['seek.com.au']
#     start_urls = ['https://www.seek.com.au/data-scientist-jobs']
    start_urls = ['https://www.seek.com.au/data-analyst-jobs']


    def parse(self, response):
        urls = response.css("a[data-automation='jobTitle']::attr(href)").extract()
        for url in urls: #keep calling parse_details
            url = response.urljoin(url)
            yield scrapy.Request(url = url, callback=self.parse_details) #callback calls the next function def parse_details

        next_page = response.css('a[data-automation="page-next"]::attr(href)').extract_first()

        if next_page is not None:
            next_page = response.urljoin(next_page)   #joins full url with the next page number
            yield scrapy.Request(next_page, callback = self.parse)

    def parse_details(self, response): 
        def description_cleaner():
            from bs4 import BeautifulSoup
            
            job_desc = response.xpath("//div[@data-automation='jobAdDetails']").get()
            job_desc = BeautifulSoup(job_desc, 'lxml').text
            return job_desc

        yield {
            'Job Title': response.css("h1[data-automation='job-detail-title']::text").get(),
            'Company': response.xpath("//span[@data-automation='advertiser-name']//text()").get(),
            'multiple_details': response.xpath('//div[@class="yvsb870 _14uh9944y o76g430"]/text()').extract(),
            'Description': description_cleaner(),
        }
        
process = CrawlerProcess()
process.crawl(JobSpider)
process.start()
```
